```python
import matplotlib.pyplot as plt
import osmnx as ox
import pandas as pd
%matplotlib inline
ox.config(log_console=True, use_cache=True)
ox.__version__
```

```python
G = ox.graph_from_place('Moscow', network_type='drive')
fig, ax = ox.plot_graph(ox.project_graph(G), node_size=3)
```

```python
# Во-первых, посмотрите на уличную сеть без сетки
# Мы можем видеть это отсутствие последовательной ориентации в его гистограмме краевых основ.
# вычислите граничные основ и визуализируйте их частоту
G = ox.add_edge_bearings(G)
bearings = pd.Series([data['bearing'] for u, v, k, data in G.edges(keys=True, data=True)])
ax = bearings.hist(bins=30, zorder=2, alpha=0.8)
xlim = ax.set_xlim(0, 360)
ax.set_title('Orinda street network edge bearings')
plt.show()
```

```python
# Распределение УДС по сторонам света
import numpy as np
n = 30
count, division = np.histogram(bearings, bins=[ang*360/n for ang in range(0,n+1)])
division = division[0:-1]
width =  2 * np.pi/n
ax = plt.subplot(111, projection='polar')
ax.set_theta_zero_location('N')
ax.set_theta_direction('clockwise')
bars = ax.bar(division * np.pi/180 - width * 0.5 , count, width=width, bottom=0.0)
ax.set_title('Orinda street network edge bearings', y=1.1)
plt.show()
```
